## Data management

### Data cleaning

```python
# Handling missing values
df.isnull().sum()
```

```python
# 1 
df["scorer"] = df["scorer"].fillna("Unknown")

# 2 
df["minute"] = df["minute"].fillna(df["minute"].median())

# 3
df["team_code"] = df["team_code"].fillna("Unknown")
df["association"] = df["association"].fillna("Unknown")

# 4
df["rank"] = df["rank"].fillna(df["rank"].median())
df["previous_rank"] = df["previous_rank"].fillna(df["previous_rank"].median())
df["points"] = df["points"].fillna(df["points"].median())
df["previous_points"] = df["previous_points"].fillna(df["previous_points"].median())
```

```python
df.isnull().sum()
```

```python
# Convertion of data types where necessary
df.dtypes
```

```python
# Removal of any duplicate entries
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

### Data wrangling

```python
# Creating new column
df.loc[:, "rank_category"] = pd.cut(
    df["rank"],
    bins=[0, 10, 50, 100, 200],
    labels=["Top 10", "Top 50", "Top 100", "Others"]
)
```

```python
df[["rank", "rank_category"]].head()
```

```python
# Combining columns
df.loc[:, "scorer_team"] = df["scorer"] + " (" + df["team"] + ")"
```

```python
df[["scorer", "team", "scorer_team"]].head()
```

```python
# Extracting data from multiple columns to create a new column
df.loc[:, "goal_timing"] = "Early Goal"
df.loc[df["minute"] >= 45, "goal_timing"] = "Late goal"
```

```python
df[["minute", "goal_timing"]].head(10)
```

## Data visualisation

```python
# Creating visualisations
```

```python
# 1 Early vs Late goals
counts = df["goal_timing"].value_counts().reindex(["Early Goal", "Late goal"])

plt.figure(figsize=(8, 3))
plt.plot(counts.values, counts.index, "o")
for i, v in enumerate(counts.values):
    plt.text(v, i, f"{v}", va="center", ha="left")

plt.xlabel("Number of Goals")
plt.ylabel("Goal Timing")
plt.title("Early vs Late Goals (Dot Plot)")
plt.grid(axis="x", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.savefig("exports/Early_vs_Late_goals.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# 2 Goals by Rank Category
plt.figure(figsize=(8, 5))
sns.countplot(x="rank_category", data=df, order=["Top 10", "Top 50", "Top 100", "Others"])
plt.title("Goals Scored by Team Rank Category")
plt.xlabel("Rank Category")
plt.ylabel("Number of Goals")
plt.savefig("exports/Goals_by_Rank_Category.png", dpi=300, bbox_inches="tight")
plt.show()
```
